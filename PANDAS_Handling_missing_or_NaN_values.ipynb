```python
#isnull() method
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df2 = df1.isnull()

print(df1.head())
print()
print(df2.head())
```

```python
#notnull() method
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df2 = df1.notnull()

print(df1.head())
print()
print(df2.head())
```

```python
#Counting the missing values in each columns
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
s = df1.isna().sum()
print(s)
```

```python
#Counting the missing values in each column with percentage
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
s = df1.isna().sum()
per = (s*100)/len(df1)
print(per)
```

```python
#Dropping rows whereb NaN values existing
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df2 = df1.dropna()
print(df2)
```

```python
#Dropping rows where NaN values existing and counting
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df2 = df1.dropna()
s = df2.isna().sum()
print(s)
```

```python
#Converting float column type into int data type
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df2 = df1.dropna()
df3 = df2.astype(int)
print(df2.head())
print()
print(df3.head())
```

```python
#Dropping NaN values by using inplace parameter
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df1.dropna(inplace=True)
print(df1)
```

```python
#Filling NaN values with zero
import pandas as pd
df1 = pd.read_csv("fruits1.csv")
df2 = df1.fillna(0)

print(df1.head())
print()
print(df2.head())
```

```python
#Filling NaN values with specific value
import pandas as pd
import numpy as np
data = [
    ["Rajan",26,40000],
    ["Daniel",16,20000],
    ["Veeru",45,90000],
    ["Venkat",np.nan,45000],
    ["Sumanth",20,95000],
    ["Shafi",np.nan,97000]
]
df1 = pd.DataFrame(data,columns=["Name","Age","Salary"])
df2 = df1.fillna(22)

print(df1)
print()
print(df2)
```

```python
#Filling NaN value with mean value
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
data = [
    ["Shahid",26,40000],
    ["Daniel",16,20000],
    ["Veeru",45,90000],
    ["Venkat",np.nan,45000],
    ["Sumanth",20,95000],
    ["Shafi",np.nan,97000]
]
df1 = pd.DataFrame(data,columns=["Name","Age","Salary"])
print(df1)
m = df1['Age'].mean()
df1['Age']=df1['Age'].fillna(m)

print()
print(df1)
```

```python
#Creating dataframe and replacing nan values with specific value
import pandas as pd
import numpy as np
data = [
    ["Shahid",26,40000],
    ["Daniel",16,20000],
    ["Veeru",45,90000],
    ["Sumanth",20,95000],
]
df1 = pd.DataFrame(data,columns=["Name","Age","Salary"])
print(df1)
df2 = df1.replace(np.nan,0)

print()
print(df2)
```
